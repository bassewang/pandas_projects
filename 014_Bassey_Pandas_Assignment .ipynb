```python
student_df = pd.read_csv('/Users/ononoinwang/Desktop/Python_Training/Pandas_Training/009_Student_data/students_complete.csv')
student_df
```

```python
school_df = pd.read_csv('/Users/ononoinwang/Desktop/Python_Training/Pandas_Training/009_Student_data/schools_complete.csv')
school_df
```

```python
complete_df = pd.merge(student_df, school_df, on='school_name')
complete_df
```

```python
stud_passing_maths = complete_df.query('math_score >=70')
stud_passing_maths
```

```python
stud_passing_reading = complete_df.query('reading_score >=70')
stud_passing_reading
```

```python
stud_passing_reading_and_math = complete_df.query('reading_score >=70 and math_score >=70')
stud_passing_reading_and_math
```

```python
stud_total_count = complete_df.groupby(['school_name']).agg(no_students = ('student_name', 'count'))
stud_total_count
```

```python
cnt_stud_passing_maths = stud_passing_maths.groupby(['school_name']).agg(no_math_pass = ('math_score', 'count'))
cnt_stud_passing_maths
```

```python
cnt_stud_passing_reading = stud_passing_reading.groupby(['school_name']).agg(no_reading_pass = ('reading_score', 'count'))
cnt_stud_passing_reading
```

```python
cnt_stud_passing_reading_and_math = stud_passing_reading_and_math.groupby(['school_name']).agg(no_passing_all = ('Student ID', 'count'))
cnt_stud_passing_reading_and_math
```

```python

total_counts = stud_total_count.join(
    (cnt_stud_passing_maths, 
    cnt_stud_passing_reading,
    cnt_stud_passing_reading_and_math
    )
)

total_counts
```

```python
total_counts['percent_maths_pass'] = round((total_counts['no_math_pass']/total_counts['no_students'])*100,2)
total_counts['percent_reading_pass'] = round((total_counts['no_reading_pass']/total_counts['no_students'])*100,2)
total_counts['percent_pass_overall'] = round((total_counts['no_passing_all']/total_counts['no_students'])*100,2)
total_counts = total_counts.sort_values(by ='percent_pass_overall', ascending=False)
total_counts = total_counts.reset_index()
total_counts
```

# 3) The Average Math Score for students of each grade level (9th, 10th, 11th, 12th) at each school 

```python
#3) The Average Math Score for students of each grade level (9th, 10th, 11th, 12th) at each school 
stud_avg_maths_score = complete_df.groupby(['school_name', 'grade']).agg(avg_maths_score = ('math_score','mean')).round(2)                                         
stud_avg_maths_score = stud_avg_maths_score.unstack().droplevel(level=0, axis=1).reset_index()
# dir(df.index.name)
# help(df.reindex)
#dir(df.columns)
stud_avg_maths_score.columns.name = ""
stud_avg_maths_score
```

# 4) The Average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school 

```python
#4) The Average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school 
stud_avg_reading_score = complete_df.groupby(['school_name', 'grade']).agg(avg_reading_score = ('reading_score','mean')).round(2)                                           
stud_avg_reading_score = stud_avg_reading_score.unstack().droplevel(level=0, axis=1).reset_index()
# dir(df.index.name)
# help(df.reindex)
#dir(df.columns)
stud_avg_reading_score.columns.name = ""
stud_avg_reading_score
```

# 5) A table that breaks down school performances based on average spending ranges (per student), and 4 reasonable bins to group school spending was created.

```python
budget_per_school = complete_df.groupby(['school_name']).agg(school_budgets = ('budget', 'mean'))
budget_per_school = budget_per_school.reset_index()
budget_per_school
```

```python
# merge total performance table [stud_total_count] with the total budgets [budget_per_school]]dataframe
# create a new calculated column - get the avg budget or spending per student in each school
# sort the "budget per student" values to understand groupings and create 4 reasonable bins to group school spending
```

```python
total_budgets_df = pd.merge(stud_total_count,budget_per_school,on='school_name')
total_budgets_df['budget_per_student'] = round((total_budgets_df['school_budgets']/total_budgets_df['no_students']), 2)
total_budgets_df = total_budgets_df.sort_values(by ='budget_per_student' , ascending=False)
total_budgets_df
```

```python
#total performance by spending/budget
#4 reasonable bins - ''< 600.0', '600.0 - 635.0', '636.0 - 645.0', '> 645.0']'
school_by_spending = pd.merge(total_counts,total_budgets_df,on='school_name', how='inner',suffixes=(' ',' '))
school_by_spending = school_by_spending.loc[:,~school_by_spending.T.duplicated(keep='first')]
school_by_spending = school_by_spending.sort_values(by ='budget_per_student', ascending=False)
school_by_spending['bins'] = pd.cut(x=school_by_spending['budget_per_student'], bins=[0, 600, 635, 645, 655], 
                    labels=['less than 600.0', 'btw 600.0 & 635.0', 'btw 636.0 & 645.0', 'greater than 645.0'])
avg_school_scores = complete_df.groupby('school_name').agg(avg_reading_scores=('reading_score', 'mean'),avg_math_scores=('math_score', 'mean'))
avg_school_scores = avg_school_scores.reset_index()
scores_by_budget = pd.merge(school_by_spending,avg_school_scores,on='school_name')
scores_per_avgspend = scores_by_budget.groupby(['bins']).agg(avg_reading_scores=('avg_reading_scores', 'mean'),
                       avg_math_scores =('avg_math_scores', 'mean'),
                       percent_reading_pass=('percent_reading_pass', 'mean'),
                       percent_maths_pass=('percent_maths_pass', 'mean'),
                       percent_pass_overall=('percent_pass_overall', 'mean')).round(3)
scores_per_avgspend = scores_per_avgspend.reset_index()
scores_per_avgspend
```

# 6) A table of Scores by School Size (small, medium, and large) - for six, you can set the definition of small, medium, and large as: less than 1k, more than 1k & less than 2k, and above 2k respectively.

```python
# 6) A table of Scores by School Size (small, medium, and large)
# small, medium, and large as: less than 1k, more than 1k & less than 2k, and above 2k respectively.
# inspect the no.of students by school to determine school size, group them into 3 buckets as per the above 
# bucket_label, names of schools under each group and then the list of maths and english score under 
# each grouping. 
```

```python
# create 3 main buckets ["Small (<1000)", "Medium (1000-2000)" , "Large (2000-5000)"]
avg_school_scores = complete_df.groupby('school_name').agg(avg_reading_scores=('reading_score', 'mean'),avg_math_scores=('math_score', 'mean'))
avg_school_scores = avg_school_scores.reset_index()
scores_by_schoolsize = pd.merge(total_counts,avg_school_scores,on='school_name')
scores_by_schoolsize['bins'] = pd.cut(x=scores_by_schoolsize['no_students'],bins = [0, 999, 1999, 5000],
                                 labels=["Small (<1000)", "Medium (1000-2000)" , "Large (2000-5000)"])
scores_per_sch_size = scores_by_schoolsize.groupby(['bins']).agg(avg_reading_scores=('avg_reading_scores', 'mean'),
                       avg_math_scores =('avg_math_scores', 'mean'),
                       percent_reading_pass=('percent_reading_pass', 'mean'),
                       percent_maths_pass=('percent_maths_pass', 'mean'),
                       percent_pass_overall=('percent_pass_overall', 'mean')).round(3)
scores_per_sch_size = scores_per_sch_size.reset_index()
scores_per_sch_size
```

# 7) A table of Scores by School Type (Charter & District) pivoted against Averge scores and % passing

```python
school_scores = pd.merge(pd.merge(total_counts,avg_school_scores,on='school_name'),school_df,on='school_name')
cols = ['no_students','no_math_pass','no_reading_pass', 'budget', 'School ID', 'size' ]
scores_by_sch_type=school_scores.drop(columns=cols)
scores_by_sch_type = scores_by_sch_type.groupby('type').agg(avg_reading_scores=('avg_reading_scores', 'mean'),
                       avg_math_scores =('avg_math_scores', 'mean'),
                       percent_reading_pass=('percent_reading_pass', 'mean'),
                       percent_maths_pass=('percent_maths_pass', 'mean'),
                       percent_pass_overall=('percent_pass_overall', 'mean')).round(3)
scores_by_sch_type = scores_by_sch_type.reset_index()
scores_by_sch_type
```
